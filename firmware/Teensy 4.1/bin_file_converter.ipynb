```python
import struct

def bin_to_candump(bin_file, txt_file):
    with open(bin_file, 'rb') as bf, open(txt_file, 'w') as tf:
        while True:
            # Read the bus identifier (1 byte), ID (4 bytes), EXT flag (1 byte), LEN (1 byte), and DATA (8 bytes)
            bus_byte = bf.read(1)
            id_bytes = bf.read(4)
            ext_flag_bytes = bf.read(1)
            len_bytes = bf.read(1)
            data_bytes = bf.read(8)

            # Check if we've reached the end of the file
            if not bus_byte or not id_bytes or not ext_flag_bytes or not len_bytes or not data_bytes:
                break

            # Unpack the values from the binary file
            bus = struct.unpack('B', bus_byte)[0]
            can_id = struct.unpack('I', id_bytes)[0]
            ext_flag = struct.unpack('B', ext_flag_bytes)[0]
            can_len = struct.unpack('B', len_bytes)[0]
            can_data = struct.unpack('8B', data_bytes)

            # Determine the bus name based on the bus identifier
            bus_name = 'can1' if bus == 1 else 'can2'

            # Format the data into candump style
            id_str = f"{can_id:08X}" if ext_flag else f"{can_id:03X}"
            data_str = ' '.join(f"{byte:02X}" for byte in can_data[:can_len])
            line = f"{bus_name} {id_str} [{can_len}] {data_str}"

            # Write the line to the text file
            tf.write(line + '\n')

# Example usage
bin_to_candump('log_001.bin', 'output.txt')
```
